- Định nghĩa hàm: Hàm được định nghĩa với bốn tham số: a, b, m, và n. Các tham số này đại diện cho giới hạn dưới a, giới hạn trên b, số hàng m, và số cột n của ma trận, tương ứng.

- Kiểm tầu vào: Hàm đầu tiên kiểm tra xem a có nhỏ hơn b hay không, và nếu m và n lớn hơn 0. Nếu điều kiện này được đáp ứng, hàm tiếp tục; nếu không, nó không thực hiện bất kỳ thao tác nào và trra đả về None.

- Tạo mảng ngẫu nhiên: Sử dụng np.random.random((m, n)) từ thư viện NumPy để tạo ra một ma trận các số ngẫu nhiên có kích thước m hàng và n cột. Mỗi phần tử của ma trận này nằm trong khoảng từ 0 đến 1.

- Tỉ lệ mảng: Tỉ lệ mảng ngẫu nhiên được tạo ra ở bước 3 để phù hợp giữa a và b. Điều này được thực hiện bằng cách nhân ma trận ngẫu nhiên với hiệu giữa b và a, và sau đó cộng a vào.

- Tính trung bình và phương sai: Tính toán trung bình kỳ vọng và phương sai của mảng được tỉ lệ bằng cách sử dụng các giới hạn được cung cấp a và b.

- In trung bình và phương sai: In trung bình và phương sai tính toán được.

- Trả về mảng tỉ lệ: Cuối cùng, hàm trả về ma trận ngẫu nhiên đã tỉ lệ.

- Gọi hàm: Hàm createMatrix() được gọi với các đối số (2, 5, 3, 4), có nghĩa là giới hạn dưới là 2, giới hạn trên là 5, số hàng là 3, và số cột là 4.

- Vì vậy, hàm sẽ tạo ra một ma trận 3x4 của các số ngẫu nhiên tỉ lệ giữa 2 và 5, in trung bình và phương sai của chúng, và trả về ma trận đã tỉ lệ.

```python
def createMatrix(a,b, m,n):
    if a < b and m > 0 and  n > 0:
        array_random = np.random.random((m, n))
        array_scaled = a + (b - a) * array_random
        excepted_mean = (a + b) / 2
        excepted_var = ((b - a) ** 2) / 12
        print("Mean:", excepted_mean)
        print("Variance:", excepted_var)
        
        return array_scaled
        

print(createMatrix(2, 5, 3, 4))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# generate 5000 numbers from a uniform distribution -3 to 3
data_random = np.random.uniform(-3, 3, 5000)
print(data_random)
```

```python
# tạo 5000 đặt trong loc to 5 and scale to 2
data_normal = np.random.normal(5, 2, 5000)
print(data_normal)
```

```python
#  vẽ biểu đồ
plt.hist(data_random, bins=np.arange(-3,3,0.25), color='c')
plt.title('Uniform Distribution')
plt.show()
```

```python
#  vẽ biểu đồ
plt.hist(data_normal, bins=np.arange(-2,13,0.5), color='c')
plt.title('Normal Distribution')
plt.show()
```
