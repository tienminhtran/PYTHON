- nh ngha hm: Hm c nh ngha vi bn tham s: a, b, m, v n. Cc tham s ny i din cho gii hn di a, gii hn trn b, s hng m, v s ct n ca ma trn, tng ng.

- Kim tu vo: Hm u tin kim tra xem a c nh hn b hay khng, v nu m v n ln hn 0. Nu iu kin ny c p ng, hm tip tc; nu khng, n khng thc hin bt k thao tc no v trra  v None.

- To mng ngu nhin: S dng np.random.random((m, n)) t th vin NumPy  to ra mt ma trn cc s ngu nhin c kch thc m hng v n ct. Mi phn t ca ma trn ny nm trong khong t 0 n 1.

- T l mng: T l mng ngu nhin c to ra  bc 3  ph hp gia a v b. iu ny c thc hin bng cch nhn ma trn ngu nhin vi hiu gia b v a, v sau  cng a vo.

- Tnh trung bnh v phng sai: Tnh ton trung bnh k vng v phng sai ca mng c t l bng cch s dng cc gii hn c cung cp a v b.

- In trung bnh v phng sai: In trung bnh v phng sai tnh ton c.

- Tr v mng t l: Cui cng, hm tr v ma trn ngu nhin  t l.

- Gi hm: Hm createMatrix() c gi vi cc i s (2, 5, 3, 4), c ngha l gii hn di l 2, gii hn trn l 5, s hng l 3, v s ct l 4.

- V vy, hm s to ra mt ma trn 3x4 ca cc s ngu nhin t l gia 2 v 5, in trung bnh v phng sai ca chng, v tr v ma trn  t l.

```python
def createMatrix(a,b, m,n):
    if a < b and m > 0 and  n > 0:
        array_random = np.random.random((m, n))
        array_scaled = a + (b - a) * array_random
        excepted_mean = (a + b) / 2
        excepted_var = ((b - a) ** 2) / 12
        print("Mean:", excepted_mean)
        print("Variance:", excepted_var)
        
        return array_scaled
        

print(createMatrix(2, 5, 3, 4))
```
