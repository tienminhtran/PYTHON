```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
data = pd.read_csv('data_spotify.csv')
data.head(10)
```

```python

# gn cho spotify_sample
spotify_sample = data.sample(1000, random_state=10000)
spotify_sample
```

```python
# chuyn t milisecond sang pht
data['duration_min'] = data['duration_ms'] / 60000
# Tnh thi gian trung bnh tnh bng pht t qun th Spotify s dng gu trc
mean_duration_data = data['duration_min'].mean()
print("Tnh thi gian trung bnh tnh bng pht t qun th Spotify s dng gu trc")
mean_duration_data
```

```python
# chuyn t milisecond sang pht
spotify_sample['duration_min'] = spotify_sample['duration_ms'] / 60000
# Tnh thi lng trung bnh tnh bng pht t Spotify_sample bng cch s dng gu trc
mean_duration_spotify = spotify_sample['duration_min'].mean()
print("Tnh thi lng trung bnh tnh bng pht t Spotify_sample bng cch s dng gu trc")
mean_duration_spotify
```

- gi tr trung bnh ca thi lng bi ht t qun th Spotify s dng gu trc l 3.856
- gi tr trung bnh ca thi lng bi ht t Spotify_sample s dng gu trc l 3.874
- cho thy gi tr spotify-sample i in cho qun th spotify nn c th dng data kiu ngu nhin  xem xt c qun th
