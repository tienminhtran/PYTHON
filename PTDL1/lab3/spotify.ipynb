```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
data = pd.read_csv('data_spotify.csv')
data.head(10)
```

```python

# gán cho spotify_sample
spotify_sample = data.sample(1000, random_state=10000)
spotify_sample
```

```python
# chuyển tù milisecond sang phút
data['duration_min'] = data['duration_ms'] / 60000
# Tính thời gian trung bình tính bằng phút từ quần thể Spotify sử dụng gấu trúc
mean_duration_data = data['duration_min'].mean()
print("Tính thời gian trung bình tính bằng phút từ quần thể Spotify sử dụng gấu trúc")
mean_duration_data
```

```python
# chuyển tù milisecond sang phút
spotify_sample['duration_min'] = spotify_sample['duration_ms'] / 60000
# Tính thời lượng trung bình tính bằng phút từ Spotify_sample bằng cách sử dụng gấu trúc
mean_duration_spotify = spotify_sample['duration_min'].mean()
print("Tính thời lượng trung bình tính bằng phút từ Spotify_sample bằng cách sử dụng gấu trúc")
mean_duration_spotify
```

### nhận xét

- giá trị trung bình của thời lượng bài hát từ quần thể Spotify sử dụng gấu trúc là 3.856
- giá trị trung bình của thời lượng bài hát từ Spotify_sample sử dụng gấu trúc là 3.874
- cho thấy giá trị spotify-sample đại điện cho quần thể spotify nên có thể dùng data kiểu ngẫu nhiên để xem xét cả quần thể

```python
# ve bieu do acousticness hist()
# Vẽ biểu đồ về độ âm thanh từ Spotify_population với các thùng có chiều rộng 0,01 từ 0 đến 1 bằng cách sử dụng pandas .hist(
plt.hist(data['acousticness'], color='blue', bins=np.arange(0, 1, 0.01))
plt.title('Acousticness')
plt.show()
```

### nhận xét

- Spotify_population: Độ âm thanh của Spotify_population có phân phối đều từ 0 đến 1
- Spotify_sample: Độ âm thanh của Spotify_sample có phân phối đều từ 0 đến 1
- các giá trị độ âm trong mẫu có thể khái quát hóa cho dân số nói chung (Vì các giá trị độ âm trong mẫu và dân số có phân phối đều từ 0 đến 1)
=> Nên các giá trị độ âm trong mẫu có thể khái quát hóa cho dân số nói chung
- Spotify_population: Độ âm thanh của Spotify_population có phân phối đều từ 0 đến 1. Điều này có nghĩa là, trong dữ liệu tổng thể, độ âm thanh có thể nằm ở bất kỳ giá trị nào từ 0 đến 1 mà không có sự thiên vị về một giá trị cụ thể nào.

- Spotify_sample: Độ âm thanh của Spotify_sample cũng có phân phối đều từ 0 đến 1. Điều này có nghĩa là, trong mẫu dữ liệu đã lấy, độ âm thanh cũng có thể nằm ở bất kỳ giá trị nào từ 0 đến 1 mà không có sự thiên vị về một giá trị cụ thể nào.

- Khái quát hóa cho dân số nói chung: Vì các giá trị độ âm trong mẫu và dân số có phân phối đều từ 0 đến 1, nên có thể khẳng định rằng mẫu dữ liệu có thể đại diện cho dân số nói chung. Điều này có nghĩa là, các phân tích và kết luận rút ra từ mẫu dữ liệu có thể được áp dụng lên tổng thể.

- Kết luận: Các giá trị độ âm trong mẫu có thể khái quát hóa cho dân số nói chung. Điều này là kết quả của việc cả mẫu và tổng thể đều có phân phối đều của độ âm thanh từ 0 đến 1.
