• Lấy tỷ lệ nhân viên theo trình độ học vấn từ attrition_pop.
• Sử dụng phương pháp lấy mẫu phân tầng theo tỷ lệ trên attrition_pop để lấy mẫu 40% của mỗi nhóm Giáo dục, đặt giai đoạn hạt giống đến năm 2022.
• Lấy tỷ lệ nhân viên theo trình độ học vấn từ attrition_strat.

```python
nv_edu = data['Education'].value_counts()
nv_edu
```

```python
#  Use proportional stratified sampling on attrition_pop to sample 40% of each Education group, setting the seed to 2022.
edu_strat = data.groupby('Education')\
.sample(frac=0.4, random_state=2022)
```

```python
edu_strat['Education'].value_counts(normalize=True)
# normalize=True là để tính phần trăm
```

```python
# Get the proportion of employees by Education level from attrition_strat.
nv_edu_strat = data['Education'].value_counts()
nv_edu_strat
```
