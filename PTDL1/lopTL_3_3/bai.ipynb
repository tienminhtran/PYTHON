Filtering for 6 country

```python
#Filtering for 6 country
top_country =["Mexico", "Guatemala", "Honduras", "Colombia", "Brazil", "Ethiopia"]
df_top = df['Country of Origin'].isin(top_country)
coffee_raing = df[df_top]
coffee_raing
```

```python
#count of simple ramdom sample
coffee_raing_sample = df.sample(frac= 0.1, random_state=2021)
coffee_raing_sample['Country of Origin'].value_counts(normalize=True)
```

```python
df['Country of Origin'].value_counts(normalize=True)
```

```python

#chọn ngẫu nhiên 10% 
coffee_raing_eq = coffee_raing_sample.groupby('Country of Origin')\
    .sample(random_state=2021)
coffee_raing_eq['Country of Origin'].value_counts(normalize=True)       
```

```python
#weighted random sampling
coffee_raing_weight = coffee_raing_sample
condition = coffee_raing_weight['Country of Origin'] == 'Mexico'
coffee_raing_weight['weight'] = np.where(condition, 2, 1)
coffee_raing_weight = coffee_raing_weight.sample(frac=1, weights='weight', random_state=2021)
coffee_raing_weight['Country of Origin'].value_counts(normalize=True)
```

```python
#varieies of coffee
varieties_pop = list(coffee_raing['Variety'].unique())
varieties_pop
```

```python
import random
varieties_sample = random.sample(varieties_pop, k = 3)
varieties_sample
```

```python
#sampling each group
variety_condition = coffee_raing['Variety'].isin(varieties_sample)
coffee_raing_cluster = coffee_raing[variety_condition].astype('category')

coffee_raing_cluster['Variety'] = coffee_raing_cluster['Variety'].cat.remove_unused_categories()
coffee_raing_cluster.groupby('Variety').sample(n=1, random_state=2021)
```

```python
#review of sampling techniques - setup
top_counted_countries = ["Mexico", "Guatemala", "Honduras", "Colombia", "Brazil", "Ethiopia"]
subset_condition = coffee_raing['Country of Origin'].isin(top_counted_countries)
coffee_raing_top = coffee_raing[subset_condition]
coffee_raing_top.shape
```

```python
#review sample random sampling
coffee_raing_srs = coffee_raing_top.sample(frac=1/3, random_state=2021)
coffee_raing_srs.shape
```

```python
 # review stratified random sampling
coffe_rating_strat = coffee_raing_top.groupby('Country of Origin').sample(frac=1/3, random_state=2021)
coffe_rating_strat.shape
```

```python
#review of cluser sampling
coffee_counries_samp = random.sample(top_counted_countries, k=0)
top_condition = coffee_raing_top['Country of Origin'].isin(coffee_counries_samp)
coffee_raing_cluster = coffee_raing_top[top_condition].astype('category')
coffee_raing_cluster['Country of Origin'] = coffee_raing_cluster['Country of Origin'].cat.remove_unused_categories()

coffee_raing_clust = coffee_raing_cluster.groupby('Country of Origin').sample(n= len(coffee_raing_top)//2)
coffee_raing_clust.shape
```
