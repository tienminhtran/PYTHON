```python
import pandas as pd
import numpy as np
```

```python
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head(5)
```

```python
netflix_subset = netflix_df[netflix_df["type"]=="Movie"]
netflix_subset
```

```python
column = ["title", "country", "genre", "release_year","duration"]
netflix_movies = netflix_subset[column]
netflix_movies
```

```python
short_movie = netflix_movies[netflix_movies["duration"] <= 60]
short_movie.head(20)
```

```python
colors = []
for data, row in netflix_movies.iterrows():
  if row["genre"] == "Children":
      colors.append("red")
  elif row["genre"] == "Documentaries":
      colors.append("blue")
  elif row["genre"] == "Stand-Up":
      colors.append("green")
  else:
      colors.append("black")
len(colors)
```

```python
import matplotlib.pyplot as plt
```

```python
print({"Childern":"Red", "Documentaries":"Blue", "Stand-Up":"Green", "Other":"Black"})
print()
fig = plt.figure(figsize=(12,8))
plt.scatter(x=netflix_movies["release_year"], y=netflix_movies["duration"], c=colors, edgecolors='none')
plt.title("Movie Duration Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.show()
```

```python
men_df = pd.read_csv("men_results.csv", index_col=0)
woman_df = pd.read_csv("women_results.csv", index_col=0)
```

```python
print(men_df.info())
print()
print(woman_df.info())
```

```python
men_results_value = men_df[(men_df['date'] > "2002-01-01") & (men_df['tournament'] == "FIFA World Cup")]
women_results_value = woman_df[(woman_df['date'] > "2002-01-01") & (woman_df['tournament'] == "FIFA World Cup")]
```

```python
men_results_value.head()
```

```python
women_results_value.head()
```

```python
men_results_value['goals_score'] = men_results_value['home_score'] + men_results_value['away_score']

women_results_value['goals_score'] = women_results_value['home_score'] + women_results_value['away_score']
```

```python
print("Men's summary")
print(men_results_value['goals_score'].describe(), "\n")
print("Standard deviation:", men_results_value['goals_score'].std(), "\n")
print("Variance:", men_results_value['goals_score'].var())

print("\n")

# Women's summary
print("Women's summary")
print(women_results_value['goals_score'].describe(), "\n")
print("Standard deviation:", women_results_value['goals_score'].std(), "\n")
print("Variance:", women_results_value['goals_score'].var())
```

```python

w = women_results_value.copy()
w["date"] =pd.to_datetime(w["date"])
w['year'] = w['date'].dt.year
w = w.groupby('year')['goals_score'].sum().reset_index()
w.plot(kind="bar",x='year')

w
```
