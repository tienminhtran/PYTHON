```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('netflix_data.csv')
df.head(5)
```

```python
netflix_subs = df[df["type"]=="Movie"]
netflix_subs
```

```python
arr_comlum = ["title", "country", "genre", "release_year","duration"]
netflix_movies = netflix_subs[arr_comlum]
netflix_movies
```

```python
sort_movie = netflix_movies[netflix_movies['duration'] <=60]
sort_movie.head(20)
```

```python
netflix_movies.genre.value_counts()
```

```python
color = []
for data, row in netflix_movies.iterrows():
    if row['genre'] == "Children":
        color.append('red')
    elif row['genre'] == "Documentaries":
        color.append('blue')
    elif row['genre'] == "Stand-Up":
        color.append('green')
        
    else:
        color.append('black')

len(color)
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], color=color, alpha=0.6)
plt.title('Netflix Movies theo năm')
plt.show()
```
