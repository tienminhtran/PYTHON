```python
import pandas as pd
import numpy as np
```

```python
# Tạo DataFrame ban đầu với chỉ mục đơn
df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_of_birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, 32, 33, 30, 31, 32]},
    index=['t1', 't2', 't3', 't4', 't5', 't6'])

# In ra DataFrame gốc với chỉ mục đơn
print("DataFrame gốc với chỉ mục đơn:")
print(df)

# Kiểm tra một giá trị có tồn tại trong chỉ mục đơn của DataFrame
print("\nKiểm tra một giá trị có tồn tại trong chỉ mục đơn của DataFrame:")
print('t1' in df.index)  # True
print('t11' in df.index)  # False

# Tạo DataFrame mới với chỉ mục đa cấp sử dụng các cột 't_id', 'school_code', và 'class'
df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_of_birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, 32, 33, 30, 31, 32],
    't_id': ['t1', 't2', 't3', 't4', 't5', 't6']})

# Đặt chỉ mục đa cấp cho DataFrame mới
df1 = df.set_index(['t_id', 'school_code', 'class'])

# In ra DataFrame mới với chỉ mục đa cấp
print(df1)

# Kiểm tra một giá trị có tồn tại trong chỉ mục đa cấp của DataFrame
print("\nKiểm tra một giá trị có tồn tại trong chỉ mục đa cấp của DataFrame:")
print('t4' in df1.index.levels[0])  # True
print('t4' in df1.index.levels[1])  # True
print('t4' in df1.index.levels[2])  # True
```

12.Viết chương trình Pandas để xây dựng một chuỗi sử dụng các cấp độ MultiIndex làm cột và chỉ mục.



```python
# Tạo mảng 2D 'sales_arrays'
sales_arrays = [['sale1', 'sale1', 'sale2', 'sale2', 'sale3', 'sale3', 'sale4', 'sale4'],
                ['city1', 'city2', 'city1', 'city2', 'city1', 'city2', 'city1', 'city2']]

# Sử dụng hàm zip để ghép các phần tử từ mảng 2D thành các tuple
sales_tuples = list(zip(*sales_arrays))

# In ra các tuple đã tạo
print("Tạo MultiIndex:")
print(sales_tuples)

# Tạo MultiIndex từ các tuple với tên là 'sale' và 'city'
sales_index = pd.MultiIndex.from_tuples(sales_tuples, names=['sale', 'city'])

# In ra MultiIndex
print(sales_index)

# Tạo một Series với giá trị ngẫu nhiên và sử dụng MultiIndex làm chỉ mục
s = pd.Series(np.random.randn(8), index=sales_index)

# In ra Series mới
print("\nXây dựng một Series sử dụng các cấp của MultiIndex:")
print(s)
```

14. Viết chương trình Pandas để trích xuất một hàng, các hàng và một giá trị cụ thể từ Khung dữ liệu cấp MultiIndex.


```python
sales_arrays = [['sale1', 'sale1', 'sale2', 'sale2', 'sale3', 'sale3', 'sale4', 'sale4'],
                ['city1', 'city2', 'city1', 'city2', 'city1', 'city2', 'city1', 'city2']]

# Ghép các phần tử từ mảng 2D thành các tuple
sales_tuples = list(zip(*sales_arrays))

# Tạo MultiIndex từ các tuple với tên là 'sale' và 'city'
sales_index = pd.MultiIndex.from_tuples(sales_tuples, names=['sale', 'city'])

# In ra các tuple đã tạo
print(sales_tuples)

# In ra MultiIndex
print("\nConstruct a Dataframe using the said MultiIndex levels:")
df = pd.DataFrame(np.random.randn(8, 5), index=sales_index)
print(df)

# Trích xuất một hàng duy nhất từ DataFrame
print("\nExtract a single row from the said dataframe:")
print(df.loc[('sale2', 'city2')])

# Trích xuất một hàng duy nhất từ DataFrame (làm lại một lần nữa)
print("\nExtract a single row from the said dataframe:")
print(df.loc[('sale2', 'city2')])

# Trích xuất nhiều hàng dựa trên một cấp của MultiIndex
print("\nExtract number of rows from the said dataframe:")
print(df.loc['sale1'])

# Trích xuất nhiều hàng dựa trên một cấp của MultiIndex (làm lại một lần nữa)
print("\nExtract number of rows from the said dataframe:")
print(df.loc['sale3'])

# Trích xuất một giá trị duy nhất từ DataFrame
print("\nExtract a single value from the said dataframe:")
print(df.loc[('sale1', 'city2'), 1])

# Trích xuất một giá trị duy nhất từ DataFrame (làm lại một lần nữa)
print("\nExtract a single value from the said dataframe:")
print(df.loc[('sale4', 'city1'), 4])
```

15. Viết chương trình Pandas để đổi tên các cột và nhãn cụ thể của Main Index của khung dữ liệu MultiIndex.


```python

sales_arrays = [['sale1', 'sale1', 'sale2', 'sale2', 'sale3', 'sale3', 'sale4', 'sale4'],
                ['city1', 'city2', 'city1', 'city2', 'city1', 'city2', 'city1', 'city2']]

# Ghép các phần tử từ mảng 2D thành các tuple
sales_tuples = list(zip(*sales_arrays))

# Tạo MultiIndex từ các tuple với tên là 'sale' và 'city'
sales_index = pd.MultiIndex.from_tuples(sales_tuples, names=['sale', 'city'])

# In ra các tuple đã tạo
print(sales_tuples)

# In ra MultiIndex
print("\nConstruct a Dataframe using the said MultiIndex levels:")
df = pd.DataFrame(np.random.randn(8, 5), index=sales_index)
print(df)

# Đổi tên các cột của DataFrame
print("\nRename the columns name of the said dataframe")
df1 = df.rename(columns={0: "col1", 1: "col2", 2:"col3", 3:"col4", 4:"col5"})
print(df1)

# Đổi tên cho các nhãn cụ thể của chỉ mục chính của DataFrame
print("\nRename specific labels of the main index of the DataFrame")
df2 = df1.rename(index={"sale2": "S2", "city2": "C2"})
print(df2)
```


16. Viết chương trình Pandas để sắp xếp MultiIndex của DataFrame. Cũng sắp xếp theo các cấp độ khác nhau của chỉ mục.

```python

# Tạo mảng dữ liệu 'sales_arrays'
sales_arrays = [['sale1', 'sale1', 'sale3', 'sale3', 'sale2', 'sale2', 'sale4', 'sale4'],
                ['city1', 'city2', 'city1', 'city2', 'city1', 'city2', 'city1', 'city2']]

# Ghép các phần tử từ mảng 2D thành các tuple
sales_tuples = list(zip(*sales_arrays))

# Tạo MultiIndex từ các tuple với tên là 'sale' và 'city'
sales_index = pd.MultiIndex.from_tuples(sales_tuples, names=['sale', 'city'])

# In ra các tuple đã tạo
print(sales_tuples)

# In ra MultiIndex
print("\nConstruct a Dataframe using the said MultiIndex levels:")
df = pd.DataFrame(np.random.randn(8, 5), index=sales_index)
print(df)

# Sắp xếp DataFrame theo MultiIndex
print("\nSort on MultiIndex DataFrame:")
df1 = df.sort_index()
print(df1)

# Sắp xếp DataFrame theo cấp chỉ mục (level) 0 của MultiIndex
print("\nSort on Index level=0 of the DataFrame:")
df2 = df.sort_index(level=0)
print(df2)

# Sắp xếp DataFrame theo cấp chỉ mục (level) 1 của MultiIndex
print("\nSort on Index level=1 of the DataFrame:")
df2 = df.sort_index(level=1)
print(df2)

# Sắp xếp DataFrame bằng cách chuyển tên của cấp chỉ mục
print("\nPass a level name to sort the DataFrame:")
df3 = df.sort_index(level="city")
print(df3)
```
