Viết chương trình Pandas để tìm vị trí của các giá trị lân cận với các giá trị nhỏ hơn ở cả hai vế trong một chuỗi cho trước.


```python
x = pd.Series([1, 8, 7, 5, 6, 5, 3, 4, 7, 1])

for i in range(len(x) - 2):
    if(x[i] < x[i + 1] and x[i + 1] > x[i + 2]):
        print("Local maximum:", x[i + 1])

for i in range(1, len(x) - 1):
    if(x[i - 1] < x[i] > x[i + 1]):
        print("Local maximum at index", i)
```

```python
# Mảng số
nums = np.array([1, 8, 7, 5, 6, 5, 3, 4, 7, 1])
# Bước 1: Tính sự khác biệt giữa các phần tử liên tiếp trong mảng `nums`
diff_1 = np.diff(nums)
# Bước 2: Tính dấu của mỗi phần tử trong kết quả của Bước 1
sign_diff_1 = np.sign(diff_1)
# Bước 3: Tính sự khác biệt giữa các phần tử liên tiếp trong mảng dấu `sign_diff_1`
diff_2 = np.diff(sign_diff_1)
# Bước 4: Tìm các vị trí trong mảng `diff_2` mà giá trị là -2
minima_positions = np.where(diff_2 == -2)[0] + 1
# In ra kết quả, tức là các chỉ mục của các điểm cực tiểu trong mảng `nums`
print("Chỉ mục của các điểm cực tiểu:", minima_positions)
```
