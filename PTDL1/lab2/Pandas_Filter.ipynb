```python

# Đọc dữ liệu từ file CSV và lưu vào DataFrame
w_a_con = pd.read_csv('world_alcohol.csv')

# Hiển thị dữ liệu mẫu từ DataFrame
print("Dữ liệu tiêu thụ rượu trên thế giới:")
print(w_a_con.head())

# Chọn các cột liên tiếp bằng tên cột (từ 'Country' đến 'Display Value')
print("\nChọn các cột liên tiếp:")
print(w_a_con.loc[:, "Country":"Display Value"].head())

# Hoặc chọn các cột liên tiếp bằng chỉ số cột (từ 2 đến 4)
print("\nLệnh thay thế:")
print(w_a_con.iloc[:, 2:5].head())

# Chọn các dòng và cột cụ thể sử dụng nhãn index
print("\nChọn các dòng có nhãn Index từ 0 đến 9 và các cột cụ thể:")
print(w_a_con.loc[0:9, ["Year", "Country", "Display Value"]])
```

```python
w_a_con = pd.read_csv('world_alcohol.csv')

# Đọc dữ liệu tiêu thụ rượu trên thế giới từ file CSV và lưu vào DataFrame mới
new_w_a_con = pd.read_csv('world_alcohol.csv')

# Hiển thị dữ liệu mẫu từ DataFrame gốc
print("Dữ liệu tiêu thụ rượu trên thế giới:")
print(w_a_con.head())

# Đổi tên tất cả các tên cột
print("\nĐổi tên tất cả các tên cột:")
w_a_con.columns = ['year', 'who_region', 'country', 'beverage_types', 'display_values']
print(w_a_con.head())

# Đổi tên chỉ một số tên cột
print("\nĐổi tên chỉ một số tên cột:")
new_w_a_con.rename(columns={"WHO region": "WHO_region", "Display Value": "Display_Value"}, inplace=True)
print(new_w_a_con.head())
```
