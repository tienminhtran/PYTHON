```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Syed Wharton'],
    'date_of_birth': ['17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age': [18.5, 21.2, 22.5, 22, 23]
})
print("Original DataFrame:")
print(df)
# sử dụng info() để xem thông tin của DataFrame
print(df.info())
# sử dụng select_dtypes() để lấy các cột có kiểu dữ liệu là số
print("\nSelect numerical columns")
print(df.select_dtypes(include = "number"))
print("\nSelect string columns")
print(df.select_dtypes(include = "object"))
```

```python

df = pd.DataFrame({
    'Name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Syed Wharton'],
    'Date_Of_Birth ': ['17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'Age': [18.5, 21.2, 22.5, 22, 23]
})

# In ra DataFrame gốc
print("DataFrame gốc:")
print(df)

# Tạo bản sao sâu của DataFrame gốc 'df' với tên là 'df1'
df1 = df.copy(deep=True)

# Loại bỏ các dòng có chỉ số 0 và 1 từ 'df'
df = df.drop([0, 1])

# Loại bỏ dòng có chỉ số 2 từ 'df1'
df1 = df1.drop([2])

# In ra các DataFrame mới sau khi loại bỏ dòng
print("\nCác DataFrame mới:")
print(df)
print(df1)

# Thực hiện merge một-một và in kết quả
print('\n"one_to_one”: kiểm tra xem các khóa merge có duy nhất trong cả hai bộ dữ liệu trái và phải:')
df_one_to_one = pd.merge(df, df1, validate="one_to_one")
print(df_one_to_one)

# Thực hiện merge một-nhiều và in kết quả
print('\n"one_to_many” hoặc “1:m”: kiểm tra xem các khóa merge có duy nhất trong bộ dữ liệu trái:')
df_one_to_many = pd.merge(df, df1, validate="one_to_many")
print(df_one_to_many)

# Thực hiện merge nhiều-một và in kết quả
print('“many_to_one” hoặc “m:1”: kiểm tra xem các khóa merge có duy nhất trong bộ dữ liệu phải:')
df_many_to_one = pd.merge(df, df1, validate="many_to_one")
print(df_many_to_one)
```

```python

df = pd.DataFrame({
    'name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Syed Wharton', 'Kierra Gentry'],
    'age': [18, 22, 85, 50, 80, 5]
})

# In ra DataFrame gốc
print("Original DataFrame:")
print(df)

# Thêm một cột mới có tên là 'age_groups', phân loại tuổi thành các nhóm 'kids', 'adult', 'elderly'
df["age_groups"] = pd.cut(df["age"], bins=[0, 18, 65, 99], labels=["kids", "adult", "elderly"])
# df["age"]: Truy cập cột "age" trong DataFrame df.

# pd.cut(...): Hàm này được sử dụng để chia các giá trị trong cột "age" thành các khoảng (bins) được xác định và gán nhãn cho mỗi khoảng.

# bins=[0, 18, 65, 99]: Xác định các khoảng tuổi, ví dụ 0-18 là nhóm "kids", 18-65 là nhóm "adult", và trên 65 là nhóm "elderly".

# labels=["kids", "adult", "elderly"]: Gán nhãn cho mỗi khoảng, nhãn này sẽ được thêm vào cột "age_groups". Ví dụ, nếu giá trị nằm trong khoảng 0-18, sẽ được gán nhãn "kids".

# df["age_groups"]: Tạo hoặc cập nhật cột mới có tên là "age_groups" trong DataFrame df với các giá trị nhãn tương ứng dựa trên khoảng tuổi.

# In ra cột mới 'age_groups'
print('\nNhóm tuổi:')
print(df["age_groups"])
```
