### ex2: Xác định thập kỷ có tỷ lệ người chiến thắng sinh ra ở Mỹ cao nhất
- Để tính tỷ lệ, trước tiên hãy tạo một cột tạo cờ cho những người chiến thắng có quốc gia sinh ra là "Hợp chủng quốc Hoa Kỳ", sau đó tạo cột thập kỷ và sử dụng cả hai để tìm tỷ lệ.

```python
nguoi_quoc_gia_USA = df[df['bornCountry'] == 'USA']
nguoi_quoc_gia_USA
nam_nhiu_nhat = nguoi_quoc_gia_USA.year.value_counts().idxmax()
nam_nhiu_nhat
nguoi_quoc_gia_USA_nam_nhiu_nhat = nguoi_quoc_gia_USA[nguoi_quoc_gia_USA['year'] == nam_nhiu_nhat]
nguoi_quoc_gia_USA_nam_nhiu_nhat
```

```python
# Tạo một cột mới có giá trị mặc định là False
df['chiến thắng'] = False

```

```python
# Tạo một Series chỉ định các chỉ số hàng thỏa mãn điều kiện từ DataFrame nguoi_quoc_gia_USA_nam_nhiu_nhat
indices_to_mark = nguoi_quoc_gia_USA_nam_nhiu_nhat.index

```

```python
# Đặt giá trị True cho các hàng có chỉ số thỏa mãn điều kiện trong cột 'chiến thắng'
df.loc[indices_to_mark, 'chiến thắng'] = True
In ra DataFrame sau khi thêm cột mới
df
```

```python
df['is_US_born'] = df['bornCountry'] == 'USA'

# Create a decade column
df['decade'] = (df['year'] // 10) * 10

# Calculate the proportion of US-born winners
proportion_us_born = df[df['is_US_born']]['decade'].value_counts(normalize=True).head(1)

proportion_us_born
```
