```python
#  Vẽ đồ thị pie-chart để hiển thị tỷ lệ dân số giữa các châu lục
# Tính tổng dân số của từng châu lục
df1['total_population'] = df1.iloc[:, 2:].sum(axis=1)

# Tạo pie-chart
plt.figure(figsize=(8, 6))
plt.pie(df1['total_population'], labels=df1['name'], autopct='%1.1f%%', startangle=140)
plt.title('Tỷ lệ dân số giữa các châu lục')
plt.axis('equal')  # Đảm bảo trục x và trục y có tỷ lệ tương đương để vẽ một hình tròn
plt.show()
```

### Câu 4: (2 điểm)

a, (1.5 điểm) Vẽ biểu đồ tăng trưởng về dân số của 10 quốc gia Đông Nam Á từ năm 2000-2020

b, (0.5 điểm) Dựa vào hình vẽ bạn hãy cho biết quốc gia nào có tốc độ tăng trưởng nhanh nhất?

```python
# Chọn dữ liệu của các quốc gia Đông Nam Á từ năm 2000 đến 2020
southeast_asian_countries = ['Vietnam', 'Thailand', 'Indonesia', 'Philippines', 'Myanmar', 'Malaysia', 'Cambodia', 'Laos', 'Singapore', 'Brunei']
df_seasia = df1[df1['name'].isin(southeast_asian_countries)]

# Chọn cột chỉ chứa dữ liệu về dân số từ năm 2000 đến 2020
years = [str(year) for year in range(2000, 2021)]
df_seasia_population = df_seasia[years]

# Tính tăng trưởng dân số của từng quốc gia
growth_rates = df_seasia_population.apply(lambda row: (row.iloc[-1] - row.iloc[0]) / row.iloc[0] * 100, axis=1)

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
for country in growth_rates.index:
    plt.plot(years, growth_rates[country], label=country)

plt.title('Tăng trưởng dân số của 10 quốc gia Đông Nam Á (2000-2020)')
plt.xlabel('Năm')
plt.ylabel('Tăng trưởng dân số (%)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
