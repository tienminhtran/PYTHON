# Vẽ biểu đồ cột biểu diễn giá trị trung bình của `LoanAmount` theo từng giới tính . Dưa ra nhận xét

```python
grbygt = df.groupby("Gender")["LoanAmount"].mean()

grbygt.plot(kind="bar", color=["red", "blue"])
plt.title("Gia tri trung binh theo gioi tinh")
plt.xlabel("Gioi tinh")
plt.ylabel("Gia tri trung binh")
plt.show()
```

- Tỉ lệ mượn nợ của Nam sẽ nhiều hơn nữ

```python
grbygt = df.groupby("Gender")["LoanAmount"].mean()

mang = np.array(grbygt)
print(mang)                                                                         
plt.bar(grbygt.index, mang, color=["red", "blue"])
plt.title("Gia tri trung binh theo gioi tinh")
plt.xlabel("Gioi tinh")
plt.ylabel("Gia tri trung binh")
plt.show()
```

Vẽ biểu đồ boxplot của cột `CoapplicantIncome` từ DataFrame df. Đặt tiêu đề biểu đồ là "Boxplot of CoapplicantIncome".. Dưa ra nhận xét
