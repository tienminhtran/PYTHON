## <u>Bài 1</u>

Green M&M Candies liên quan đến Dataset 18 trong file excel. Tìm tỉ lệ mẫu của M&M có màu xanh lá cây (green). Sử dụng kết quả để xây dựng 1 ước lượng khoảng tin cậy 95% của tổng thể M&M có màu xanh.

```python
df = pd.read_excel('18_M_M.xls')
df
```

```python
# Green M&M Candies liên quan đến Dataset 18 trong file excel. Tìm tỉ lệ mẫu của M&M có màu xanh lá cây (green). Sử dụng kết quả để xây dựng 1 ước lượng khoảng tin cậy 95% của tổng thể M&M có màu xanh.

# Tính tỉ lệ mẫu của M&M có màu xanh lá cây (green)
n = df.size


tb = df['Green'].mean()

dlc = df['Green'].std()

f = df['Green'].size/n

print('Kich thuoc mau: ', n)
print('Trung binh mau: ', tb)
print('Do lech chuan mau: ', dlc)
print('Ty le GREEN: ', f)
```

```python
# UOC LUONG KHOANG TIN CAY 95% CUA TONG THE M&M CO MAU XANH
alpha = 0.05

z = st.norm.ppf(1-alpha/2)

ci = [tb - z*dlc/math.sqrt(n), tb + z*dlc/math.sqrt(n)]

ci
```

```python
green_ratios = df['Green']

# Tính tỉ lệ trung bình và độ lệch chuẩn
mean_green_ratio = np.mean(green_ratios)
std_dev_green_ratio = np.std(green_ratios)

print("Tỉ lệ trung bình của M&M có màu xanh:", mean_green_ratio)
print("Độ lệch chuẩn của tỉ lệ mẫu:", std_dev_green_ratio)

# Số lượng mẫu
n = len(green_ratios)

# Tính khoảng tin cậy 95% sử dụng phân phối chuẩn
alpha = 0.05  # Mức ý nghĩa
z_critical = st.norm.ppf(1 - alpha/2)  # Đối với mức ý nghĩa 0.05

margin_of_error = z_critical * (std_dev_green_ratio / np.sqrt(n))

confidence_interval = (mean_green_ratio - margin_of_error, mean_green_ratio + margin_of_error)

print("Khoảng tin cậy 95% của tỉ lệ mẫu của M&M có màu xanh:", confidence_interval)
```

b. Xây dựng ước lượng khoảng tin cậy 95% về tỉ lệ phần trăm các sinh viên cao đẳng tăng cân trong năm thứ 1.


```python
# loc df2['WTAPR'] > df2['WTSEP']
tong_can_nang = df2.loc[df2['WTAPR'] > df2['WTSEP']].shape[0]
print(tong_can_nang) # xet de dung pp chuan
tong_sv = df2.shape[0]

ty_le = tong_can_nang/tong_sv

dlc = math.sqrt(ty_le*(1-ty_le)/tong_sv)

khoang_tin_cay = st.norm.interval(0.95, loc=ty_le, scale=dlc)

print("a.tìm ước lượng điểm tốt nhất của tỉ lệ phần trăm các sinh viên cao đẳng tăng cân trong năm thứ 1", ty_le)

print("b.khoảng tin cậy 95% về tỉ lệ phần trăm các sinh viên cao đẳng tăng cân trong năm thứ 1: ", khoang_tin_cay)
```
