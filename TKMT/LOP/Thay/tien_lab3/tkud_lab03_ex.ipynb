## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random as rd
````

````python
x = rd.randint(1000, size=100)
print(x)
````

````python
print("10 Phan tu dau tien: ", x[:10])
print("5 Phan tu cuoi cung: ", x[-5:])
a = []
for i in x:
    if i % 5 == 0 and i > 100:
        a.append(i)
print("Cac phan tu chia het cho 5 va lon hon 100: ", a)
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
a = []
for i in x:
    if i < 100:
        a.append(i)
print("Cac phan tu nho 100", a)
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
print("Tong cac phan tu: ",np.sum(x))
print("Tim gia trung binh: ",x.mean())
print("Gia tri trung vi",np.median(x)) 

# Tìm Q1, Q2, Q3, IQR, min, max
print("Q1: ",np.quantile(x, 0.25))
print("Q2: ",np.quantile(x, 0.5))
print("Q3: ",np.quantile(x, 0.75))
print("IQR: ",np.quantile(x, 0.75)-np.quantile(x, 0.25))
print("Min: ",np.min(x))
print("Max: ",np.max(x))
# tim phan vi 20%
print("Phan vi 20%: ",np.quantile(x,20/100))
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
for i in range(5):
    print("Lan ", i, rd.choice(x,size=10,replace=False))
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python
max_count = 0
max_value = None

for value in x:
    count = 0
    for item in x:
        if item == value:
            count += 1
    if count > max_count:
        max_count = count
        max_value = value
print(max_value)
````

## SEABORN

import seaborn as sns

Seaborn là một thư viện hỗ trợ trực quan hóa được viết dựa trên thư viện `matplotlib`. Trong bài tập này, bạn hãy tự tìm hiểu thư viện `seaborn` và trực quan hóa dữ liệu tiền tip tại một nhà hàng.

Bạn có thể dùng đoạn mã sau để đọc dữ liệu có sẵn trong `seaborn`

```python
import seaborn as sns

df = sns.load_dataset("tips")

```
Hoặc bạn có thể đọc dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy lựa chọn loại đồ thị phù hợp để trả lời những câu hỏi sau:

````python
import pandas as pd
import numpy as np
````

````python
import seaborn as sns
````

````python
data = pd.read_csv("tips.csv")
data.head()
````

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
# tinh tong tien tip cot sun
tong_tien = data.groupby('day')['tip'].sum()

# Vẽ đồ thị thể hiện tổng tiền tip theo ngày
plt.bar(tong_tien.index, tong_tien.values)

````

### 2. Vẽ đồ thị thể hiện tổng tiền tip theo thời gian dùng bữa (time)


````python
a = data.groupby('time')['tip'].sum()
plt.bar(a.index, a.values)
````

### 3. Vẽ đồ thị histogram của tổng hoá đơn


````python
plt.hist(data['total_bill'], bins=30, color="green")
````

### 4. Vẽ đồ thị boxplot của tổng hoá đơn


````python
plt.boxplot(data['total_bill'])
plt.title("Boxplot của total_bill")
````

### 5. Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần


````python
data.boxplot(column='total_bill', by='day')
plt.title("Boxplot của total_bill theo ngày")
````

### 6. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip. (`scatter plot`)


````python
plt.scatter(data['total_bill'], data['tip'])
plt.title('Mối quan hệ giữa tổng hóa đơn và tiền tip')
plt.xlabel('Tổng hóa đơn (USD)')
plt.ylabel('Tiền tip (USD)')
plt.show()
````

### 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?


````python
## 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg')
plt.show()
# Nhận xét
# Từ biểu đồ trên, ta thấy rằng tổng hóa đơn và tiền tip có mối tương quan dương với nhau. Tức là khi tổng hóa đơn tăng thì tiền tip cũng tăng theo. Điều này cũng phản ánh qua biểu đồ jointplot ở trên.
````

### 8. Tìm hiểu và sử dụng hàm `pairplot` trong seaborn để trực quan bộ dữ liệu. Theo bạn, trong bộ dữ liệu có những thuộc tính nào liên quan với nhau không?



````python
sns.pairplot(data)
plt.show()
# nhận xét
#  nhận xét pairplot
````

### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`



````python
tong, con = plt.subplots(1,2)
### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`


#  tổng hóa đơn theo thời gian dùng bữa Lunch 
a = data[data['time'] == 'Lunch']
con[0].hist(a['total_bill'], bins=30, color="green")
con[0].set_title('Lunch')



b = data[data['time'] == 'Dinner']
con[1].hist(b['total_bill'], bins=30, color="red")
con[1].set_title('Dinner')
plt.show()
````

### 10. Vẽ hai đồ thị boxplot cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`

````python
tong, con = plt.subplots(1,2)
### 10. Vẽ boxplot để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`
a = data[data['time'] == 'Lunch']
con[0].boxplot(a['total_bill'])
con[0].set_title('Lunch')

b = data[data['time'] == 'Dinner']
con[1].boxplot(b['total_bill'])
con[1].set_title('Dinner')

plt.show()
````
