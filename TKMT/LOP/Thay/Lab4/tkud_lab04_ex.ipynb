**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
lamda = 3
prob_1_call = poisson.pmf(1, lamda)
print("Xác suất để có 1 cuộc gọi trong 1 phút là:", prob_1_call)
prob_2_call = poisson.pmf(2, lamda)
print("Xác suất để có 2 cuộc gọi trong 1 phút là:", prob_2_call)
prob_3_call = poisson.pmf(3, lamda)
print("Xác suất để có 3 cuộc gọi trong 1 phút là:", prob_3_call)
```

**Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại.

b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
mu = 500
sigma = np.sqrt(16)  # do lech chaun = sqrt( phương sai)
z1 = (505 - mu) / sigma
z2_lower = (495 - mu) / sigma
z2_upper = (505 - mu) / sigma
z3 = (495 - mu) / sigma

# Tính xác suất từ z-score
p1 = 1 - norm.cdf(z1)
p2 = norm.cdf(z2_upper) - norm.cdf(z2_lower)
p3 = norm.cdf(z3)

# Tính tỷ lệ của mỗi loại
total_probability = p1 + p2 + p3
ratio1 = p1 / total_probability
ratio2 = p2 / total_probability
ratio3 = p3 / total_probability

print("Tỷ lệ của mỗi loại:")
print("Loại 1:", ratio1)
print("Loại 2:", ratio2)
print("Loại 3:", ratio3)

# b. Mô phỏng lấy mẫu 20 trái cây
sample_20 = np.random.normal(mu, sigma, 20)


plt.hist(sample_20, bins=20, edgecolor='black')
plt.title('Histogram of 20 fruits')
plt.xlabel('Trọng lượng')
plt.ylabel('SL')
plt.show()


# c. Mô phỏng lấy mẫu 100 trái cây
sample_100 = np.random.normal(mu, sigma, 100)

plt.hist(sample_100, bins=20, edgecolor='black')
plt.title('Histogram of 100 fruits')
plt.xlabel('Trọng lượng')
plt.ylabel('SL')
plt.show()

mean_100 = np.mean(sample_100)
std_100 = np.std(sample_100)
print("Trung bình trọng lượng của 100 trái cây:", mean_100)
print("Độ lệch chuẩn của trọng lượng của 100 trái cây:", std_100)

print("Nhận xét:")
print("Khi lấy mẫu 20 trái cây và 100 trái cây, giá trị trung bình và độ lệch chuẩn có xu hướng tiến gần đến giá trị trung bình và độ lệch chuẩn của toàn bộ quần thể.") 
```
