1. Viết chương trình nhập vào số nguyên n (n>0). In ra các số nguyên tố từ 1..n


```python
def soNguyenTo(n):
    if n < 2:
        return 0
    else:
        for i in range(2, n):
            if n % i == 0:
                return 0
        return 1    
    
n = int(input("Nhap n: "))
for i in range(1, n+1):
    if soNguyenTo(i):
        print(i)
```

4. Viết chương trình yêu cầu người dùng nhập họ tên. Hiển thị tên người dùng sau khi đã chuẩn hóa. Biết rằng tên đã chuẩn hóa là:
- Ký tự đầu viết hoa
- Không có khoảng trắng thừa
- Không có ký tự đặc biệt


```python

ho_ten = input("Nhập họ và tên của bạn: ")
ten_chuan_hoa = ho_ten.strip().title()
ten_chuan_hoa = ''.join(e for e in ten_chuan_hoa if e.isalpha()) 
print("Tên của bạn sau khi chuẩn hóa là:", ten_chuan_hoa)
```

5. Viết chương trình nhập vào một số nguyên dương (n>0). Cho biết n có phải là số đối xứng hay không? Biết số đối xứng là số viết xuôi hay viết ngược đều như nhau

```python
def soDoiXung(n):
    if n == n[::-1]:
        return 1
    else:
        return 0   
    
n = input("Nhap n: ")
if soDoiXung(n):
    print("So doi xung")
else:
    print("Khong doi xung")
```
