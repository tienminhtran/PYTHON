```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv('babies.txt', sep='\s+')
df_1.head()
```

```python
# Chia Thành Hai Nhóm
# hut thuoc
me_hut = df_1[df_1['smoke']==1]
# không hút thuốc
me_ko = df_1[df_1['smoke']==0]

tong,con = plt.subplots(1,2)
con[0].boxplot(me_hut['bwt'])
con[0].set_title('Mẹ hút thuốc')

con[1].boxplot(me_ko['bwt'])
con[1].set_title('Mẹ không hút thuốc')
tong.suptitle('Trọng lượng sinh sản phẩm của mẹ hút thuốc và không hút thuốc')
```
