```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('penguins-data.csv')
```

```python
# tính tỉ lệ % của "Torgersen" trong cột "island
tam = df['island'].value_counts()['Torgersen']
print((tam*100)/df['island'].count())
```

```python
# bill_depth_mm,flipper_length_mm
print(df['bill_depth_mm'].mean())
print(df['bill_depth_mm'].median())
print(df['bill_depth_mm'].std())


print(df['flipper_length_mm'].median())
print(df['flipper_length_mm'].std())
print(df['flipper_length_mm'].median())
```

```python
# Lấy 5 giá trị lớn nhất của cột "body_mass_g", chỉ lấy giá trị
tam = df.nlargest(5, 'body_mass_g')
tam['body_mass_g'].to_list()
```

```python
# Lấy 5 giá trị nhỏ nhất của cột "body_mass_g", chỉ lấy giá trị
tam = df.nsmallest(5, 'body_mass_g')
tam['body_mass_g'].to_list()
```

```python
# Lấy 5 giá trị nhỏ nhất của cột "body_mass_g", lấy thông tin
tam = df.nsmallest(5, 'body_mass_g')
tam 
```

```python
# Biểu đồ phân phối displot, Phân Phối Độ Dài Mỏ (bill_length_mm) Theo Loài và Giới Tính
sns.displot(data=df, x="bill_length_mm", hue="species", col="sex", kind="kde", fill=True)
plt.title('Phân Phối Độ Dài Mỏ Theo Loài và Giới Tính')
plt.show()
```

```python
# 2/ 	Biểu đồ scatterplot: Để thể hiện mối quan hệ giữa cân nặng (body_mass_g) và chiều dài mỏ (bill_length_mm),  cho tất cả các mẫu . 
plt.scatter(df['body_mass_g'], df['bill_length_mm'])
plt.title('Phân Phối Độ Dài Mỏ Theo Loài và Giới Tính')
plt.show()
```

```python
# 3/ 	Biểu đồ cột bar, Để so sánh trung bình cân nặng của chim cánh cụt Adelie theo loài và giới tính
plt.bar(df['species'], df['body_mass_g'])
plt.title('Trung Bình Cân Nặng Theo Loài và Giới Tính')
plt.show()
```
